# Siguiente Clase
1. de la fuente df_codes vamos a identificar los codigos generales

aglomerado nivel 2

```python
df_parent = df_codes[df_codes['Level']== 2]
df_parent
```

2. Crear la dimesion de codigos se llamara dim_codes

```python
#limpiar de datos nulos la columna code_comm
#El dato crudo nunca se toca
df_codes_clean = df_codes.copy()
df_codes_clean = df_codes_clean[df_codes_clean['Code_comm'].notnull()]
df_codes_clean
```

```python
# funcion que permite extraer el codigo de la partida particulas y separarla de la general
#[:numero] toma los primeros digitos que se indiquen [::numero] toma los ultimos digitos que se indiquen
def clean_code(text):
    text = str(text)
    parent_code = None
    if len(text) == 11:
        code = text[:5]
        parent_code = text[:1]
    else:
        code = text[:6]
        parent_code = text[:2]
    try:
        parent = df_parent[df_parent['Code_comm']== parent_code]['Description'].values[0]
    except:
        parent = None
    return (code,parent)
```

Crear en el dataframe df_codes_clean los campos de codigo y descripcion general

```python
df_codes_clean[['code_clean','parent_description']]= df_codes_clean.apply(lambda x: clean_code(x['Code']), axis= 1, result_type='expand')
df_codes_clean
```

```python
#construir la estructura de dim_codes
df_codes_clean = df_codes_clean[df_codes_clean['code_clean'].notnull()][['code_clean','Description', 'parent_description']]
df_codes_clean.rename(columns={'code_clean':'code','Description':'description'},inplace=True)
df_codes_clean
```

```python
dim_codes = df_codes_clean.copy()
dim_codes.reset_index(drop= True,inplace =True)
dim_codes['id_code'] = dim_codes.index+1
dim_codes['code']= dim_codes['code'].astype('int64')
dim_codes = dim_codes[['id_code','code','description','parent_description']]
dim_codes.info()
```
