# Siguiente Clase
1. de la fuente df_codes vamos a identificar los codigos generales
detalle nivel 1
aglomerado nivel 2

```python
df_parent = df_codes[df_codes['Level']== 2]
df_parent
```

2. Crear la dimesion de codigos se llamara dim_codes

```python
#limpiar de datos nulos la columna code_comm
#El dato crudo nunca se toca
df_codes_clean = df_codes.copy()
df_codes_clean = df_codes_clean[df_codes_clean['Code_comm'].notnull()]
df_codes_clean
```

```python
# funcion que permite extraer el codigo de la partida particulas y separarla de la general
#[:numero] toma los primeros digitos que se indiquen [::numero] toma los ultimos digitos que se indiquen
def clean_code(text):
    text = str(text)
    parent_code = None
    if len(text) == 11:
        code = text[:5]
        parent_code = text[:1]
    else:
        code = text[:6]
        parent_code = text[:2]
    try:
        parent = df_parent[df_parent['Code_comm']== parent_code]['Description'].values[0]
    except:
        parent = None
    return (code,parent)
```

Crear en el dataframe df_codes_clean los campos de codigo y descripcion general

```python
df_codes_clean[['code_clean','parent_description']]= df_codes_clean.apply(lambda x: clean_code(x['Code']), axis= 1, result_type='expand')
df_codes_clean
```

```python
#construir la estructura de dim_codes
df_codes_clean = df_codes_clean[df_codes_clean['code_clean'].notnull()][['code_clean','Description', 'parent_description']]
df_codes_clean.rename(columns={'code_clean':'code','Description':'description'},inplace=True)
df_codes_clean
```

```python
dim_codes = df_codes_clean.copy()
dim_codes.reset_index(drop= True,inplace =True)
dim_codes['id_code'] = dim_codes.index+1
dim_codes['code']= dim_codes['code'].astype('int64')
dim_codes = dim_codes[['id_code','code','description','parent_description']]
dim_codes.info()
```

```python
dim_codes
```

## Crear dimension dim_countries

```python
#limpiar el data frame df_country
df_country_clean = df_country.copy()
df_country_clean = df_country_clean[['alpha-3','country','region','sub-region']]
df_country_clean = df_country_clean[df_country_clean['alpha-3'].notnull()]
df_country_clean.rename(columns={'alpha-3':'country_code'},inplace=True)

dim_countries=df_country_clean.copy()
dim_countries.reset_index(drop = True, inplace=True)
dim_countries['id_countries']= dim_countries.index+1
dim_countries= dim_countries[['id_countries','country_code','country','region','sub-region']]
dim_countries
```

4. Construir la dimensiones dim_quantity,dim_time,dim_flow

```python
#funcion que permite crear de una columna una dimension de un solo valor
def create_dimension(data,id_name):
    list_keys = []
    value=1
    for i in data:
        list_keys.append(value)
        value= value +1
    return pd.DataFrame({id_name:list_keys,'values':data})
```

```python
#para dim_quantity
dim_quantity = create_dimension(df_trades['quantity_name'].unique(),'id_quantity')
dim_quantity
```

```python
#para dim_time
dim_time = create_dimension(df_trades['year'].unique(),'id_time')
dim_time
```

```python
#para dim_flow
dim_flow = create_dimension(df_trades['flow'].unique(),'id_flow')
dim_flow
```

5. Construir la tabla de hechos

```python
df_fact_trades = df_trades.copy()
df_fact_trades.dropna(subset = ['country_code','year','comm_code','flow','quantity_name'],how='any',inplace=True)
df_fact_trades
```

```python
df_fact_trades =[] #reinicia la tabla para que no se dupliquen las columnas de los joins
df_fact_trades = df_trades.copy()
df_fact_trades.dropna(subset = ['country_code','year','comm_code','flow','quantity_name'],how='any',inplace=True)

df_fact_trades = df_fact_trades.merge(dim_codes[['code','id_code']],how='left',left_on='comm_code',right_on='code')
df_fact_trades = df_fact_trades.merge(dim_countries[['country_code','id_countries']],how='left',left_on='country_code',right_on='country_code')
df_fact_trades = df_fact_trades.merge(dim_quantity,how='left',left_on='quantity_name',right_on='values')
df_fact_trades = df_fact_trades.merge(dim_time,how='left',left_on='year',right_on='values')
df_fact_trades = df_fact_trades.merge(dim_flow,how='left',left_on='flow',right_on='values')  #aqui no se especifican las columnas porque la tabla solo tiene las 2 columnas necesarias




df_fact_trades
```

```python
df_fact_trades.reset_index(drop=True,inplace=True)
df_fact_trades['id_trades'] = df_fact_trades.index+1
fact_trades = df_fact_trades[['id_trades','id_code','id_countries','id_time','id_flow','id_quantity','trade_usd','kg','quantity']]
fact_trades.info
```

```python
dim_quantity.rename(columns={'values':'quantity_name'},inplace=True)
dim_flow.rename(columns={'values':'flow'},inplace=True)
dim_time.rename(columns={'values':'year'},inplace=True)
dim_quantity.info()
print('------------------------------------------------------')
dim_flow.info()
print('------------------------------------------------------')
dim_time.info()
```
